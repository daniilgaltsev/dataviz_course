```python
import altair as alt
import pandas as pd
import numpy as np
```

```python
np.random.seed(1103)
x = np.random.normal(size=(100,))
y = x * 2.3 - 0.2 + np.random.normal(size=(100,))
x.shape, y.shape
```

```python
df = pd.DataFrame(dict(x=x, y=y))
df.head()
```

```python
alt.Chart(df).mark_circle().encode(
    x="x",
    y="y"
).interactive()
```

```python
df["feat1"] = np.random.randint(0, 3, 100)
df["feat1"] = df["feat1"].astype(str)
df["feat2"] = df.x > 0
df.head()
```

```python
alt.Chart(df).mark_circle().encode(
    x="x",
    y="y",
    color="feat1",
    tooltip=["feat2"]
).interactive()
```
